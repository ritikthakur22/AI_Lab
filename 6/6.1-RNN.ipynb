# 6.1. RNN

Suppose, the cook at the college canteen has a rule. I am supposing the cook is a "he". 
1. So, he cooks `MoMos` when it is `sunny`, and 
2. He cooks `Chowmein ` when it is `raining`. 

So, Sunny=>Momo's and Rainy=>Chowmein. 

This can be models using a simple neural network (NN).

Using one-hot encoding the inputs and the outputs. 

|  Sunny | Rainy |
|  ----------- | ----------- |
|  1 | 0 |
| 0 | 1|

And,
|  Momos | Chowmein |
|  ----------- | ----------- |
|  1 | 0 |
|  0 | 1|


```python
import torch
import torch.nn as nn

# Linear model: 2 inputs → 2 outputs
model = nn.Linear(in_features=2, out_features=2, bias=False)

# Initialize weight matrix manually
with torch.no_grad():
    model.weight.copy_(torch.tensor([
        [1.0, 0],
        [0,  1.0]
    ]))

# Example input (batch of 1 sample)
x1 = torch.tensor([[0.0, 1]]) # rainy
x2 = torch.tensor([[1.0, 0.0]]) # sunny

for x in [x1,x2]:
    y = model(x) 
    print(y)
```

Here, we can model this as a Neural Network, because we already know the connection between the input and the output. 

We know that the cook's choice of dish depends on the weather.

The equation for RNN is as follows: 

Hidden Layer: 
$$
\boxed{
h_t
=
\tanh
\left(
W_{ih}x_t
+
W_{hh}h_{t-1}
+
b_{ih}
+
b_{hh}
\right)}
$$

and, Output Layer:

$$
\boxed{
y_t = W_{fc}h_t + b_{fc}} 
$$


Where, 

- $x_t$ = input at time step $t$
- $h_t$ = hidden state at time step $t$
- $y_t$ = output logits at time step $t$
- $W_{ih}$ = input-to-hidden matrix
- $W_{hh}$ = hidden-to-hidden matrix
- $W_{fc}$ = hidden-to-output matrix

![RNN](https://colah.github.io/posts/2015-08-Understanding-LSTMs/img/RNN-unrolled.png)

Input vector:

$$
x_t =
\begin{bmatrix}
A\\
B\\
C\\
Sunny\\
Rainy
\end{bmatrix}
$$

For example:

Dish = A, Weather = Rainy

$$
x_t =
\begin{bmatrix}
1\\
0\\
0\\
0\\
1
\end{bmatrix}
$$
