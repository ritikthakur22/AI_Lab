## 7.1. Variable Length of input and output

A Recurrent Neural Network, or `RNN`, is a network that operates on a sequence and uses its own output as input for subsequent steps.

In Sequence prediction with a single `RNN`,
- Every input corresponds to an output

This works well when there is a one-to-one alignment between inputs and outputs, such as:
- Part-of-speech tagging
- Named entity recognition
- Frame-by-frame video labeling


**Part-of-Speech (POS) Tagging**:

Assigns a **grammatical label** (e.g., noun, verb, adjective) to each word in a sentence.

Example:
```
Input : The    cat    sat    on    the    mat
Output: DT     NN     VBD    IN    DT     NN
```
Where, 
- `DT` → Determiner (article or word that specifies a noun)
- `NN` → Noun (person, place, thing, or idea)
- `VBD` → Verb, past tense
- `IN` → Preposition (shows a relationship, such as location)
- `WRB` → Wh-adverb (used to ask about a place, time, reason, or manner; e.g., where)

**Named Entity Recognition (NER)**
Identifies and classifies **named entities** such as people, organizations, and locations in a sentence.

Example:
```
Input : Steve  Jobs   founded  Apple   in   California
Output: B-PER  I-PER  O        B-ORG   O    B-LOC
```
Where, 
- `B-PER` → Beginning of a person's name
- `I-PER` → Inside/continuation of a person's name
- `B-ORG` → Beginning of an organization name
- `B-LOC` → Beginning of a location name
- `O` → Not a named entity


> In both tasks, **every input word corresponds to exactly one output label**, making them suitable for a simple RNN.

### 7.1.1. The problem 

However, this cause an issue in problems like: 
- `Machine translation` (an English sentence may translate into a longer or shorter French sentence).
- `Text summarization` (a long document becomes a short summary).
- `Speech recognition` (many audio frames map to a much shorter sequence of words).
- `Image captioning` (a single image produces a sequence of words).

In these tasks:
- There is not necessarily one output per input.
- The model often needs to read the entire input sequence before generating outputs.
- Output length is usually variable.

Let's take an example of Machine Translation: 

| # | Input Sequence | Output Sequence |
|---|----------------|-----------------|
| 1 | `English`: I am hungry. | `Nepali`: मलाई भोक लागेको छ। |
| 2 | `English`: I ate. | `German`: Ich habe gegessen. |
| 3 | `English`: I will become the Pirate King! | `Japanese`: 海賊王に俺はなる！ (*Kaizoku-ō ni ore wa naru!*) |

Or examples where one language expresses an idea with a single word while another needs a phrase:

| # | Input Sequence | Output Sequence |
|---|----------------|-----------------|
| 4 | `German`: Übermorgen | `English`: the day after tomorrow |
| 5 | `German`: Vorgestern | `English`: the day before yesterday |
| 6 | `German`: Schadenfreude | `English`: pleasure derived from another person's misfortune |
| 7 | `Japanese`: 木漏れ日 (*Komorebi*) | `English`: sunlight filtering through the leaves of trees |
| 8 | `Japanese`: 積読 (*Tsundoku*) | `English`: buying books and never reading them |
| 9 | `Japanese`: よろしく (*Yoroshiku*) | `English`: please take care of me / I look forward to working with you |
| 10 | `Japanese`: 頑張る (*Ganbaru*) | `English`: do your best / keep going / hang in there |
