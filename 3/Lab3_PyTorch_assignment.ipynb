### Tensors in PyTorch
A tensor is a multi-dimensional array used to store and manipulate data. Tensors are similar to NumPy arrays but can run on GPUs for faster computation.

### **3.1. Tensor Dimensions**
- **Scalar (0-D Tensor):** A single number, such as 5 or -3.14.
- **Vector (1-D Tensor):** A one-dimensional array, such as [1, 2, 3, 4].
- **Matrix (2-D Tensor):** A two-dimensional array, like a table with rows and columns: [[1, 2], [3, 4]].
- **Tensor (N-D Tensor):** The generalized algebraic framework for multi-dimensional numerical arrays of an arbitrary number of axes (N>3). Tensors handle complex, high-dimensional structures natively, such as color images.

#### **3.1.4.4. Tensor datatypes**

- PyTorch supports multiple tensor data types for storing numerical data.
- The default data type is torch.float32, commonly used in deep learning.
- Floating-point data types include:
  - torch.float16 (16-bit)
  - torch.float32 (32-bit)
  - torch.float64 (64-bit)
- Integer data types are available in 8-bit, 16-bit, 32-bit, and 64-bit formats.
- Integer types store whole numbers, while floating-point types can store decimal values.
- Different data types provide different levels of precision and memory usage.
- Higher precision offers better numerical accuracy but requires more memory and computation.
- Lower precision reduces memory consumption and increases computational speed.
- Choosing an appropriate data type helps balance performance and accuracy in machine learning applications.

```python
# Default datatype for tensors is float32
float_32_tensor = torch.tensor([2.0, 4.0, 6.0],
                               dtype=None, # defaults to None, which is torch.float32 or whatever datatype is passed
                               device=None, # defaults to None, which uses the default tensor type
                               requires_grad=False) # if True, operations performed on the tensor are recorded 

float_32_tensor.shape, float_32_tensor.dtype, float_32_tensor.device
```

```python
float_16_tensor = torch.tensor([3.0, 6.0, 9.0],
                               dtype=torch.float16) # torch.half would also work

float_16_tensor.dtype
```

## **3.3. Tensor Operations**

Tensor operations include arithmetic operations, matrix multiplication, aggregation, reshaping, and broadcasting. These operations form the foundation of deep learning computations and numerical processing in PyTorch.

In deep learning, different types of data such as images, text, audio, videos, and other information are converted into **tensors** (multi-dimensional arrays of numbers).

A model learns by repeatedly performing mathematical operations on these tensors to discover patterns and relationships in the data. During training, millions of calculations may be carried out.

Some of the most common tensor operations include:

- Addition
- Subtraction
- Multiplication (element-wise)
- Division
- Matrix multiplication

By combining these operations, (neural) networks/models can learn useful representations of data and make predictions.

```python
# Create a tensor of values and add a number to it
tensor = torch.tensor([1, 2, 3])
tensor + 10
```

```python
# Multiply it by 10
tensor * 10
```

```python
tensor
```

```python
# Subtract and reassign
tensor = tensor - 10
tensor
```

```python
# Can also use torch functions
torch.multiply(tensor, 10)
```

```python
tensor
```

```python
tensor = tensor + 10
# Element-wise multiplication (each element multiplies its equivalent, index 0->0, 1->1, 2->2)
print(tensor, "*", tensor)
print("Equals:", tensor * tensor)
print("Equals:", torch.multiply(tensor, tensor))
```

```python
# Let's create a tensor and perform element-wise multiplication and matrix multiplication on it.
tensor = torch.tensor([1, 2, 3])

# Element-wise matrix multiplication
print(tensor * tensor)

# Matrix multiplication
print(torch.matmul(tensor, tensor))
```

```python
tensor_A = torch.tensor([[1, 2, 3],
                         [4, 5, 6]])

tensor_B = torch.tensor([[7, 8, 9],
                         [10, 11, 12]])

# The operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n")
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n")

output = torch.matmul(tensor_A, tensor_B.T)

print("Output:\n")
print(output)
print(f"\nOutput shape: {output.shape}")
```

#### **Reshaping**

- `torch.reshape(input, shape)`:	Reshapes input to shape (if compatible), can also use torch.Tensor.reshape().

```python
# Create a tensor
import torch
x = torch.arange(1., 9.)
print(x, x.shape)

# Add an extra dimension
x_reshaped = x.reshape(1, 8) # try other shapes, it will produce an error.
x_reshaped, x_reshaped.shape
```

#### **Stack**

- `torch.stack(tensors, dim=0)`:	Concatenates a sequence of tensors along a new dimension (dim), all tensors must be same size.

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0) 
x_stacked
```

```python
# Stacking using dim=1
x_stacked = torch.stack([x, x, x, x], dim=1) 
x_stacked
```

#### **Squeeze and Unsqueeze funtions.**

```python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")

## Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

#### **Permute**

- `torch.permute()` is used to rearrange the order of a tensor's dimensions (axes) without changing the underlying data values.

```python
import torch

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(x.shape)

y = x.permute(1, 0)

print(y)
print(y.shape)
```

## 3.4. Indexing


- Indexing allows access to specific elements, rows, columns, or slices of a tensor. 
- It is useful for data extraction, manipulation, filtering, and selecting subsets of information from large datasets.

```python
# Create a tensor 
import torch
x = torch.arange(1, 10).reshape(1, 3, 3)
x, x.shape

# Let's index bracket by bracket
print(f"First square bracket:\n{x[0]}") 
print(f"Second square bracket: {x[0][0]}") 
print(f"Third square bracket: {x[0][0][0]}")
```

```python
print(x[:, 0]) # Get all values of 0th dimension and the 0 index of 1st dimension
print(x[:, :, 1])# Get all values of 0th & 1st dimensions but only index 1 of 2nd dimension
print(x[:, 1, 1]) # Get all values of the 0 dimension but only the 1 index value of the 1st and 2nd dimension
print(x[0, 0, :]) # same as x[0][0]
```

## 3.5. Pytorch Tensor to NumPy arrays and vice versa

The two main methods you'll want to use for NumPy to PyTorch (and back again) are:

- `torch.from_numpy(ndarray)` - NumPy array -> PyTorch tensor.
- `torch.Tensor.numpy()` - PyTorch tensor -> NumPy array.

```python
# NumPy array to tensor
import torch
import numpy as np
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7) # create a tensor of ones with dtype=float32
numpy_tensor = tensor.numpy() # will be dtype=float32 unless changed
tensor, numpy_tensor
```

## 3.6. Tensors on GPU

- GPUs provide parallel processing capabilities that significantly accelerate tensor computations. 
- PyTorch allows tensors to be moved between CPU and GPU devices, enabling faster training and inference for deep learning models.

```python
!nvidia-smi
```

```python
# Check for GPU
import torch
torch.cuda.is_available()
```

```python
# Set device type
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
# Count number of devices
torch.cuda.device_count()
```

```python
# Create tensor (default on CPU)
tensor = torch.tensor([1, 2, 3])

# Tensor not on GPU
print(tensor, tensor.device)

# Move tensor to GPU (if available)
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

```python
# Instead, copy the tensor back to cpu
tensor_back_on_cpu = tensor_on_gpu.cpu().numpy()
tensor_back_on_cpu
```
